```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('/content/flats - flats.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.duplicated().sum()
```

```python
df.isnull().sum()
```

```python
df.drop(columns=['link','property_id'],inplace=True)
```

```python
df.rename(columns={'area':'price_per_sqft'},inplace=True)
df.head()
```

```python
df['society'].value_counts()
```

```python
df['society'] = (
    df['society']
    .astype(str)
    .str.replace(r'\d+(\.\d+)?\s?★', '', regex=True)
    .str.strip()
    .str.lower()
)
df['society'].value_counts()
```

```python
df[df['price'] == 'Price on Request'].shape[0]
```

```python
df=df[df['price'] != 'Price on Request']
```

```python
df['price'].value_counts()
```

```python
df=df[df['price'].str.split(" ").str.len()!=1]
```

```python
## same scale
def price_scaling(x):
  if(type(x)==float):
    return x
  else:
    if(x[1]=='Lac'):
      return round(float(x[0])/100,2)
    else:
      return round(float(x[0]),2)

df['price']=df['price'].str.split(" ").apply(price_scaling)
df.head(5)
```

```python
df['price_per_sqft'].value_counts()
```

```python
df['price_per_sqft']=df['price_per_sqft'].str.split('/').str.get(0).str.replace('₹','').str.replace(',','').str.strip().astype(float)
df.head(5)
```

```python
df['bedRoom'].value_counts()
```

```python
df=df[~df['bedRoom'].isnull()]  # remove nan value
df['bedRoom']=df['bedRoom'].str.split(' ').str.get(0).astype(int)
df.head()
```

```python
df['bathroom'].value_counts()
```

```python
df['bathroom'].isnull().sum()
```

```python
df['bathroom'] = df['bathroom'].str.split(' ').str.get(0).astype(int)
df['bathroom']
```

```python
df['balcony'].value_counts()
```

```python
df['balcony'].isnull().sum()
```

```python
df['balcony']=df['balcony'].str.split(' ').str.get(0).replace('No','0')
df['balcony'].value_counts()
```

```python
df['additionalRoom'].value_counts()
```

```python
df['additionalRoom'].isnull().sum()
```

```python
df['additionalRoom']=df['additionalRoom'].fillna("not available")
df['additionalRoom'] = df['additionalRoom'].str.lower().str.strip()
df['additionalRoom'].value_counts()
```

```python
df['floorNum'].value_counts()
```

```python
df['floorNum'].isnull().sum()
```

```python
df=df[~df['floorNum'].isnull()] ## reomove nan value
df['floorNum']=df['floorNum'].str.lower().str.split(" ").str.get(0).str.replace('ground','0').str.replace('lower','0').str.replace('basement','-1').str.extract(r'(\d+)')
df['floorNum'].value_counts()
```

```python

df['facing'].value_counts()
```

```python
df['facing'].isnull().sum()
```

```python
df['facing'].fillna('NA',inplace=True)
df['facing'].value_counts()
```

```python
## new feature
df['area']=round((df['price']*10000000)/df['price_per_sqft'])
df['property_type']='flat'
df.head()
```

```python
df.info()
```

```python
df.to_csv('flats_cleaned.csv',index=False)
```
